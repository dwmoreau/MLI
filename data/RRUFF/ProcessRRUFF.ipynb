```python
import os

from RRUFFEntry import get_sample_names
from RRUFFEntry import RRUFFEntry
```

```python
RRUFF_directory = '/Users/DWMoreau/MLI/data/RRUFF/'
sample_names = get_sample_names(RRUFF_directory)
```

```python
good = 0
for sample_index, sample_name in enumerate(sample_names):
    #print(f'{sample_name} {sample_index}')
    rruff_entry = RRUFFEntry(RRUFF_directory, sample_name, False)
    if rruff_entry.status is False:
        continue

    rruff_entry.process_dif_file()
    if rruff_entry.status is False:
        rruff_entry.export()
        continue

    rruff_entry.process_dif_file()
    if rruff_entry.status is False:
        rruff_entry.export()
        continue

    rruff_entry.setup_entry()
    if rruff_entry.status is False:
        rruff_entry.export()
        continue
    rruff_entry.export()
    good += 1
print(good, len(sample_names))
```
