```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
url = 'https://rruff.info/all/display=default/'
res = requests.get(url)
htmlData = res.content
parsed_home_page = BeautifulSoup(htmlData, 'html.parser')
anchors = parsed_home_page.find_all("a", class_="page_link_2")
```

```python
skip = ['R050548']
start_at = 'R060044'
```

```python
process_entries = False
for a in anchors:
    entry_page_url = url + a['href']
    entry_name = entry_page_url.split('/')[-1]
    if entry_name == start_at:
        process_entries = True
    if process_entries and not entry_name in skip:
        print(entry_name)
        parsed_entry_page = BeautifulSoup(requests.get(entry_page_url).content, 'html.parser')
        has_diffraction = False
        for table in parsed_entry_page.find_all('table'):
            table_head = table.find_all('th', class_='sample_detail_section_head')
            if len(table_head) > 0 and table_head[0].string:
                if 'POWDER DIFFRACTION' in table_head[0].string:
                    xray_table = table
                    for td in xray_table.find_all('td'):
                        if td.string and 'Powder' in td.string:
                            has_diffraction = True
        if has_diffraction:
            print('   Processing')
            entry_dir = os.path.join('RRUFF', entry_name)
            if not os.path.exists(entry_dir):
                os.mkdir(entry_dir)
            found_processed = False
            found_raw = False
            found_dif = False
            for a2 in xray_table.find_all('a'):
                if 'Processed' in a2['href'] and not found_processed:
                    processed_file_url = a2['href']
                    processed_file_name = processed_file_url.split('/')[-1]
                    processed_file = requests.get(processed_file_url)
                    with open(os.path.join(entry_dir, processed_file_name), 'w') as out_file:
                        out_file.write(processed_file.text)
                    found_processed = True
                """
                if 'RAW' in a2['href'] and not found_raw:
                    raw_file_url = a2['href']
                    raw_file_name = raw_file_url.split('/')[-1]
                    raw_file = requests.get(raw_file_url)
                    with open(os.path.join(entry_dir, raw_file_name), 'w') as out_file:
                        out_file.write(raw_file.text)
                    found_raw = True
                """
                if 'DIF' in a2['href'] and not found_dif:
                    diff_file_url = a2['href']
                    diff_file_name = diff_file_url.split('/')[-1]
                    diff_file = requests.get(diff_file_url)
                    with open(os.path.join(entry_dir, diff_file_name), 'w') as out_file:
                        out_file.write(diff_file.text)
                    found_dif = True
        
```
